Summary and Insights:
1. Doesn't contain data of NewYork.
2. Cities with more than 1000 accidents/year are very less, is around 4.24% of the total cities
3. Around 1110 cities have reported only one accident.
4. The number of accidents per city decreases monotonically.
5. A high percentage of accidents occur between 2PM-5PM (probably people in a hurry to get to home), next highest is 6AM-8AM.
6. On weekends no of accidents are less. It is maximum on Friday.
7. Trends of accidents on Monday and Sunday are different, as expected.
8. Towards the end of the year, accidents are more frequent. More accidents happen during winter than in summer.
